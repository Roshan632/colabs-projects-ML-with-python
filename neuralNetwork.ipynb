```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=2.5
Y=1.25
```

```python
w=np.random.randn(1)
print('Initially,w=',w)
```

```python
y_out = X*w
print(y_out)
```

```python
epochs =20
lr=0.01
errorList=[]
for epoch in range(epochs):
  y_out=X*w

  error = np.square(Y-y_out)   #E=(Y-y_out)^2   E={y-(w*X)}^2
  errorList.append(error)
  gradient = -2*(Y-y_out)*X

  w=w - gradient*lr

```

```python
plt.plot(range(epochs),errorList)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error vs Epochs')
plt.show()
```
