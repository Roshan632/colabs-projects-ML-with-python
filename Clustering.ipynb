```python
X=df[['petal length (cm)','petal width (cm)']]
```

```python
X
```

```python
X.head()
```

```python


Error=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
  kmeans.fit(X)
  Error.append(kmeans.inertia_)
plt.plot(range(1,11),Error)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Error')
plt.show()

```
