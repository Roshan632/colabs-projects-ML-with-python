```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
#dataset
m=0.7   # Yp zati dida ni hunxa value
c=0.3   # Yp zati dida ni hunxa value

X=np.arange(0,1,0.02) #in 1 there is 0.02 step

Y=m*X+c

print("Shape of X=",X.shape)
print("Shape of Y=",Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
def plot_data(X_train=X_train,Y_train=Y_train,X_test=X_test,Y_test=Y_test,predictions=None):
    plt.figure(figsize=(8,6))
    plt.scatter(X_train,Y_train,label="Training Data",s=10,color="blue")
    plt.scatter(X_test,Y_test,label="Testing Data",s=10,color="red")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Scatter Plot of Training and Testing Data :X vs Y")

    if predictions is not None:
        plt.scatter(X_test,predictions,label="Predicted data",s=10,color="green")
    plt.legend()

```

```python
plot_data()
```

```python
# Initialze weights and Biases

w= np.random.randn(1)
b= np.random.randn(1)

print("w=",w)
print("b=",b)
```

```python
Y_out=X_test*w+b
plot_data(predictions=Y_out)
```

```python
def loss_fn(Y_true,Y_pred):
    loss=np.sum(np.square(Y_true-Y_pred))/len(Y_true)
    return loss
```

```python
# Training

epochs=200
lr=0.1
trainLossList=[]
testLossList=[]
for epoch in range(epochs):

    #Forward Pass
    Y_pred=X_train*w+b

    #Compute Train Loss
    trainLoss=loss_fn(Y_train,Y_pred)
    trainLossList.append(trainLoss)

    #Compute the Gradients  slope
    gradientW=(-2*np.sum(X_train*(Y_train-Y_pred))/len(Y_train))
    gradientB=(-2*np.sum(Y_train-Y_pred))/len(Y_train)

    #Backpropagate
    #Update the Parameters
    w=w-lr*gradientW
    b=b-lr*gradientB

    #Compute Test Loss
    test_pred_test=X_test*w+b
    testLoss=loss_fn(Y_test,test_pred_test)
    testLossList.append(testLoss)

    if(epoch%20==0):
        print("Epoch=",epoch,"Train Loss=",trainLoss,"Test Loss=",testLoss)
```

```python
Y_pred=X_test*w+b
plot_data(predictions=Y_pred)
```
