```python
import pandas as pd
from sklearn.model_selection import train_test_split,cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
import numpy as np

from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=load_iris()
data
```

```python
X=data.data
Y=data.target
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape

```

```python
dt= DecisionTreeClassifier()
rf=RandomForestClassifier()
knn=KNeighborsClassifier()
```

```python
print(f'dt score is {dt.score(X_test,Y_test)}')
print(f'rf score is {rf.score(X_test,Y_test)}')
print(f'knn score is {knn.score(X_test,Y_test)}')
```

```python
def Calculate_score(model):
  model.fit(X_train,Y_train)
  return model.score(X_test,Y_test)
```

```python
print(f"\nDecision Tree Score: {Calculate_score(dt)}")
print(f"\nRandom Forest Score: {Calculate_score(rf)}")
print(f"\nK neighbor Score: {Calculate_score(knn)}")
```

```python
# prompt: #Research kaggle and take raw data classification type dealing null value mean max scalar which moddel is best find ans pred and evalute

# Since the provided code uses the Iris dataset which has no missing values,
# we will simulate missing values and demonstrate handling them with mean imputation.

# Simulate some missing values
for col in range(X_train.shape[1]):
    missing_indices = np.random.choice(X_train.shape[0], size=int(X_train.shape[0] * 0.1), replace=False)
    X_train[missing_indices, col] = np.nan

# Impute missing values using the mean
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test) # Use the imputer fitted on training data

# Since the Iris dataset features are already on a similar scale, scaling is not strictly necessary
# for the models used (Decision Tree, Random Forest, K-Nearest Neighbors).
# However, for completeness and demonstrating the process, we'll add it.

# Scale the features (using StandardScaler)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# Retrain and evaluate models with imputed and scaled data
dt_scaled = DecisionTreeClassifier()
rf_scaled = RandomForestClassifier()
knn_scaled = KNeighborsClassifier()

dt_scaled.fit(X_train_scaled, Y_train)
rf_scaled.fit(X_train_scaled, Y_train)
knn_scaled.fit(X_train_scaled, Y_train)

print("\nScores after imputation and scaling:")
print(f'dt_scaled score is {dt_scaled.score(X_train_scaled, Y_train):.4f}')
print(f'rf_scaled score is {rf_scaled.score(X_train_scaled, Y_train):.4f}')
print(f'knn_scaled score is {knn_scaled.score(X_train_scaled, Y_train):.4f}')
print(f'dt_scaled score is {dt_scaled.score(X_test_scaled, Y_test):.4f}')
print(f'rf_scaled score is {rf_scaled.score(X_test_scaled, Y_test):.4f}')
print(f'knn_scaled score is {knn_scaled.score(X_test_scaled, Y_test):.4f}')

# Evaluate models with cross-validation on imputed and scaled data
def Calculate_score_scaled(model):
  model.fit(X_train_scaled, Y_train)
  return model.score(X_test_scaled, Y_test)

print(f"\nDecision Tree Score (Scaled): {Calculate_score_scaled(dt_scaled):.4f}")
print(f"\nRandom Forest Score (Scaled): {Calculate_score_scaled(rf_scaled):.4f}")
print(f"\nK neighbor Score (Scaled): {Calculate_score_scaled(knn_scaled):.4f}")

# Cross-validation on imputed and scaled data
print("\nCross-validation scores (Scaled):")
print("Decision Tree:", cross_val_score(DecisionTreeClassifier(), X_train_scaled, Y_train, cv=5).mean())
print("Random Forest:", cross_val_score(RandomForestClassifier(), X_train_scaled, Y_train, cv=5).mean())
print("K-Nearest Neighbors:", cross_val_score(KNeighborsClassifier(), X_train_scaled, Y_train, cv=5).mean())

# GridSearchCV on imputed and scaled data (for Decision Tree as in previous code)
gs_scaled = GridSearchCV(
    DecisionTreeClassifier(),
    {'criterion': ['gini', 'entropy'], 'max_depth': [1, 2, 3, 4, 5]}
)
gs_scaled.fit(X_train_scaled, Y_train)

print("\nBest parameters for Decision Tree after scaling and imputation:", gs_scaled.best_params_)
print("Best cross-validation score for Decision Tree after scaling and imputation:", gs_scaled.best_score_)

# Find the best model among the trained models based on test set accuracy
models = {'Decision Tree': dt_scaled, 'Random Forest': rf_scaled, 'K-Nearest Neighbors': knn_scaled}
best_model_name = None
best_accuracy = 0

print("\nIndividual model accuracies on the test set (after imputation and scaling):")
for name, model in models.items():
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(Y_test, y_pred)
    print(f"{name}: {accuracy:.4f}")
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model_name = name

print(f"\nBased on test set accuracy, the best performing model is: {best_model_name} with accuracy {best_accuracy:.4f}")

# Make predictions using the best model
if best_model_name is not None:
    best_model = models[best_model_name]
    predictions = best_model.predict(X_test_scaled)
    print("\nPredictions using the best model:")
    print(predictions)

    # Evaluate the best model more comprehensively
    from sklearn.metrics import classification_report, confusion_matrix
    print(f"\nEvaluation of the best model ({best_model_name}):")
    print("Confusion Matrix:")
    print(confusion_matrix(Y_test, predictions))
    print("\nClassification Report:")
    print(classification_report(Y_test, predictions))

```
